**Regressão logística**

Pergunta do projeto: *Quais fatores familiares, educacionais, raciais e de perfil demográfico mais influenciam a probabilidade de estar empregado na população do Nordeste brasileiro?*

```python
import statsmodels.api as sm
import pandas as pd
import numpy as np

# Cria variável alvo binária: 1 para empregado, 0 para desempregado
df_nordeste['empregado'] = (df_nordeste['condicao_trabalho'] == 1).astype(int)

# Seleciona variáveis explicativas
X_cols = ['sexo', 'cor_raca', 'arranjo_familiar', 'grau_parentesco']
continuous_cols = ['idade', 'anos_estudo']

X_df_prep = df_nordeste[X_cols + continuous_cols].copy()

# Convert integer-coded categorical variables to 'category' dtype before get_dummies
# This ensures get_dummies treats them as distinct categories, not continuous numbers.
for col in X_cols:
    X_df_prep[col] = X_df_prep[col].astype('category')

# Gera dummies para categóricas
X = pd.get_dummies(X_df_prep, drop_first=True)

# Explicitly convert all columns in X to float to ensure compatibility with statsmodels
X = X.astype(float)

# --- Handle multicollinearity: Remove columns with zero variance ---
# Columns with zero standard deviation indicate no variation, leading to singular matrix
constant_columns = X.columns[X.std() == 0]
if not constant_columns.empty:
    print(f"Dropping constant columns due to zero variance: {list(constant_columns)}")
    X = X.drop(columns=constant_columns)

y = df_nordeste['empregado']

# Ajusta modelo de regressão logística
# sm.add_constant(X) adds an intercept term
# Increased maxiter for better convergence
modelo = sm.Logit(y, sm.add_constant(X)).fit(maxiter=1000)
print(modelo.summary())
```

Reajustando a base de dados para facilitar a interpretação

1- Agrupando categorias menos frequentes

```python
# Para grau_parentesco (mantendo 01, 02, 03, 04; outros viram 'outros')
freq = df_nordeste['grau_parentesco'].value_counts()
pouco_frequente = freq[freq < 2000].index  # Ajuste o número conforme a amostra
df_nordeste['grau_parentesco_agrup'] = df_nordeste['grau_parentesco'].replace(list(pouco_frequente), 'outros')

# Repitindo o processo para arranjo_familiar
freq_af = df_nordeste['arranjo_familiar'].value_counts()
pouco_frequente_af = freq_af[freq_af < 2000].index
df_nordeste['arranjo_familiar_agrup'] = df_nordeste['arranjo_familiar'].replace(list(pouco_frequente_af), 'outros')
```

2- Revendo variáveis explicativas

```python
X_cols = ['sexo', 'cor_raca', 'idade', 'anos_estudo', 'arranjo_familiar_agrup', 'grau_parentesco_agrup']
X_df_prep = df_nordeste[X_cols].copy()

for col in ['sexo', 'cor_raca', 'arranjo_familiar_agrup', 'grau_parentesco_agrup']:
    X_df_prep[col] = X_df_prep[col].astype('category')

X = pd.get_dummies(X_df_prep, drop_first=True)

# Explicitly convert all columns in X to float to ensure compatibility with statsmodels
X = X.astype(float)

y = df_nordeste['empregado']

modelo = sm.Logit(y, sm.add_constant(X)).fit(maxiter=500)
print(modelo.summary())
```

 **Interpretação**

```python
import numpy as np
import pandas as pd

# Extrai coeficientes e calcula odds ratio
results = modelo.summary2().tables[1]

results['odds_ratio'] = np.exp(results['Coef.'])

# Seleciona apenas as colunas desejadas para exibição
tabela_resumo = results[['Coef.', 'odds_ratio', 'P>|z|']]
print(tabela_resumo)
```
